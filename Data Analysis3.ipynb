#### 1. The memory usage of the data is around 6.1 mb.How can we reduce the memory usage of the data set?


To reduce the memory usage:

**1. Dropping the Missing Values:**
   * Dropping the missing values can reduce the memory usage, if we try to fill the missing values that will lead to the incorrect analysis and interpretations.

**2. Dropping the Duplicate Values:**
   * Dropping the duplicate values can be useful for the accurate analysis and to remove the redundancy.

**3. Type Casting:**
   * Using the appropriate data types for the columns can reduce the memory usage.We have to Choose the smallest possible data type that can represent the data accurately.

**4. Dropping the Unwanted Columns:**
   * Deleting the unnamed columns and the columns that are not useful for the analysis can reduce memory size.
   
**5. Dealing with the Outliers:**
   * Replacing the outliers or removing the outliers can reduce the memory usage.

#### 2. What is the Average price of vehicle by fuel type and gearbox type.Give a plot


```python
new_df = pd.DataFrame(df.groupby(["fuelType","gearbox"])["price"].mean())
```

```python
sns.barplot(data= new_df, x="fuelType", y="price", hue="gearbox")
```

#### conclusion:
* The fuel type elektro with gearbox type automatik price is very high compared to others.
* The fuel type hybrid with gearbox type manuell price is very high compared to others.
* The fuel type lpg with gearbox type automatik and manuell price is less compared to others.

#### 3. What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot


```python
power_new= pd.DataFrame(df.groupby(["vehicleType","gearbox"])["powerPS"].mean())
```

```python
sns.barplot(data= power_new, x="vehicleType", y= "powerPS", hue="gearbox")
```
